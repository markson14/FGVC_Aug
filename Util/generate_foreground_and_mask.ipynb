## Configurations

We'll be using a model trained on the MS-COCO dataset. The configurations of this model are in the ```CocoConfig``` class in ```coco.py```.

For inferencing, modify the configurations a bit to fit the task. To do so, sub-class the ```CocoConfig``` class and override the attributes you need to change.

````python
class InferenceConfig(coco.CocoConfig):
    # Set batch size to 1 since we'll be running inference on
    # one image at a time. Batch size = GPU_COUNT * IMAGES_PER_GPU
    GPU_COUNT = 1
    IMAGES_PER_GPU = 1

config = InferenceConfig()
config.display()
````

## Create Model and Load Trained Weights

````python
# Create model object in inference mode.
model = modellib.MaskRCNN(mode="inference", model_dir=MODEL_DIR, config=config)

# Load weights trained on MS-COCO
model.load_weights(COCO_MODEL_PATH, by_name=True)
````

## Class Names

The model classifies objects and returns class IDs, which are integer value that identify each class. Some datasets assign integer values to their classes and some don't. For example, in the MS-COCO dataset, the 'person' class is 1 and 'teddy bear' is 88. The IDs are often sequential, but not always. The COCO dataset, for example, has classes associated with class IDs 70 and 72, but not 71.

To improve consistency, and to support training on data from multiple sources at the same time, our ```Dataset``` class assigns it's own sequential integer IDs to each class. For example, if you load the COCO dataset using our ```Dataset``` class, the 'person' class would get class ID = 1 (just like COCO) and the 'teddy bear' class is 78 (different from COCO). Keep that in mind when mapping class IDs to class names.

To get the list of class names, you'd load the dataset and then use the ```class_names``` property like this.
```
# Load COCO dataset
dataset = coco.CocoDataset()
dataset.load_coco(COCO_DIR, "train")
dataset.prepare()

# Print class names
print(dataset.class_names)
```

We don't want to require you to download the COCO dataset just to run this demo, so we're including the list of class names below. The index of the class name in the list represent its ID (first class is 0, second is 1, third is 2, ...etc.)

## Generating foreground object and mask image

````python
IMAGE_DIR = r'D:\Resources\Inat_Partial\Aves_Small_SS1'
MASK_DIR = r'D:\Resources\Inat_Partial\Aves_Small_SS1_Mask'
FOREGROUND_DIR = r'D:\Resources\Inat_Partial\Aves_Small_SS1_Object'
min_object_area = 128 * 128
min_score = 0.7
max_mask_roi_ratio = 0.3
bird_class_id = 15

for subdir_name in os.listdir(IMAGE_DIR):
    print('Generating',subdir_name)
    subdir_path = os.path.join(IMAGE_DIR, subdir_name)
    mask_sub_path = os.path.join(MASK_DIR, subdir_name)
    foreground_sub_path = os.path.join(FOREGROUND_DIR, subdir_name)
    if os.path.isdir(mask_sub_path) == False:
        os.mkdir(mask_sub_path)
    if os.path.isdir(foreground_sub_path) == False:
        os.mkdir(foreground_sub_path)
        
    for image_name in os.listdir(subdir_path):
        image = skimage.io.imread(os.path.join(subdir_path, image_name))
        image_area = image.shape[0] * image.shape[1]
        results = model.detect([image], verbose=0)
        r = results[0]
        
        # Create mask and Save binary image of mask
        mask = np.zeros(image.shape[:2])
        mask_count = 0
        for roi, class_id, scores in zip(r['rois'], r['class_ids'], r['scores']):
            roi_area = (roi[2] - roi[0]) * (roi[3] - roi[1])
            # make sure the roi is not loo large, the generated background would be bad otherwise
            if roi_area/image_area > max_mask_roi_ratio:
                continue
            mask[roi[0]:roi[2],roi[1]:roi[3]] = 255
            # Valid mask if it is a bird
            if class_id == bird_class_id and score >= min_score:
                mask_count += 1
        if mask_count > 0:
            _ = cv2.imwrite(os.path.join(mask_sub_path, image_name), mask)
        
        image_name_no_ext = image_name.split('.')[0]
        # Create transparent images of the objects
        for index, (roi, class_id, score) in enumerate(zip(r['rois'], r['class_ids'], r['scores'])):
            roi_area = (roi[2] - roi[0]) * (roi[3] - roi[1])
            # Make sure that the object is a bird and that neither the roi and score are not too small
            if (class_id != bird_class_id or 
                (roi_area < min_object_area) or 
                score < min_score):
                continue
            obj = np.zeros(image.shape[:2] + (4,))
            # set the bgr value based on the mask (opencv requirements)
            #obj[:,:,0:3] = image * r['masks'][:,:,index][:,:,np.newaxis]
            obj[:,:,0] = image[:,:,2] * r['masks'][:,:,index]
            obj[:,:,1] = image[:,:,1] * r['masks'][:,:,index]
            obj[:,:,2] = image[:,:,0] * r['masks'][:,:,index]
            # set the alpha based on the mask (note that alpha has to be 0 or 255)
            obj[:,:,3] = obj[:,:,3] + r['masks'][:,:,index] * 255
            # now crop the image based on the roi
            obj = obj[roi[0]:roi[2],roi[1]:roi[3],:]
            _ = cv2.imwrite(os.path.join(foreground_sub_path, image_name_no_ext+'_'+str(index)+'.png'), obj)
````
